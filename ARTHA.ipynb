```python
import matplotlib.pyplot as plt
import numpy as np

# Set up the figure with a clean, wide aspect ratio
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), facecolor='white')

# --- Chart 1: Coverage ---
categories_1 = ['Traditional\nBureau Score', 'Single-Source\nAlt-Data Score', 'ARTHA\n(Multi-Source)']
values_1 = [22, 51, 84]
colors_1 = ['#BDBDBD', '#FFA000', '#004D40'] # Grey, Orange, Deep Teal

bars_1 = ax1.bar(categories_1, values_1, color=colors_1, width=0.5)

# Clean up Chart 1
ax1.set_title('Credit-Invisible MSME Coverage', fontsize=14, fontweight='bold', color='#1A1A1A', pad=20)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['left'].set_visible(False)
ax1.get_yaxis().set_ticks([]) # Remove y-axis numbers

# Add data labels directly above bars
for bar in bars_1:
    height = bar.get_height()
    ax1.annotate(f'{height}%',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 5),  # 5 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom', fontsize=13, fontweight='bold', color=bar.get_facecolor())

# --- Chart 2: Turnaround ---
categories_2 = ['Manual Document-\nBased Review', 'ARTHA Real-Time\nAPI Scoring']
# Representing roughly to scale visually, but keeping the actual text labels accurate
values_2 = [6.5, 0.15]
colors_2 = ['#BDBDBD', '#E64A19'] # Grey, Deep Orange

bars_2 = ax2.barh(categories_2, values_2, color=colors_2, height=0.4)

# Clean up Chart 2
ax2.set_title('Credit Decision Turnaround', fontsize=14, fontweight='bold', color='#1A1A1A', pad=20)
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['bottom'].set_visible(False)
ax2.get_xaxis().set_ticks([]) # Remove x-axis numbers

# Add data labels directly next to horizontal bars
labels_2 = ['6.5 days', '~215 min']
for i, bar in enumerate(bars_2):
    width = bar.get_width()
    ax2.annotate(labels_2[i],
                xy=(width, bar.get_y() + bar.get_height() / 2),
                xytext=(5, 0),  # 5 points horizontal offset
                textcoords="offset points",
                ha='left', va='center', fontsize=13, fontweight='bold', color=bar.get_facecolor())

plt.tight_layout()
plt.show()
# Use plt.savefig('elegant_charts.png', dpi=300, transparent=True) to save it!
```
